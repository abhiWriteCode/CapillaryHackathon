```python
import numpy as np
import pandas as pd
from scipy import spatial
import operator
```

```python
user_cluster = pd.read_csv('Derived/Clustered_Data.csv')
indexes = pd.read_csv('Derived/ClusterIndexes.csv')
```

```python
cluster_frequency = pd.read_csv('Derived/modified_EachCluster.csv')
```

```python
user_cluster.drop(columns=['Predicted'], inplace=True)
```

```python
vector_dict = user_cluster.set_index('productid').T.to_dict('list')
cluster_dict = indexes.set_index('ClusterNo').T.to_dict('list')
```

```python
def maxoutcomes(cluster_id, product_id):
    distances = []
    
    for i, val in enumerate(cluster_dict[cluster_id]):
        try:
            if val != product_id:
                distances.append(1 - spatial.distance.cosine(vector_dict[product_id], vector_dict[val])) #11139524
                
        except:
            pass
    
    index, value = max(enumerate(distances), key=operator.itemgetter(1))
    return cluster_dict[cluster_id][index]
```

```python
maxoutcomes(36, 11139192)
```

```python
arr = []
cl = cluster_frequency.copy()
array = cl.values
```

```python
arr = []
for i, val in enumerate(array):
    pos, j = max(enumerate(val), key=operator.itemgetter(1))
    
    arr.append(cluster_frequency.columns[pos])
    array[i, pos] = 0
cluster_frequency['Top4'] = pd.Series(arr)
cluster_frequency
```

```python
cluster_frequency = cluster_frequency[['Top1', 'Top2', 'Top3', 'Top4']]
cluster_frequency
```
