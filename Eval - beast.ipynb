```python
import numpy as np
import pandas as pd
from scipy import spatial
import operator
```

```python
user_cluster = pd.read_csv('Analytics/Clustered_Data.csv')
indexes = pd.read_csv('Analytics/ClusterIndexes.csv')
```

```python
user_cluster.drop(columns=['Predicted'], inplace=True)
```

```python
vector_dict = user_cluster.set_index('productid').T.to_dict('list')
cluster_dict = indexes.set_index('ClusterNo').T.to_dict('list')
```

```python
def maxoutcomes(cluster_id, product_id):
    distances = []
    
    for i, val in enumerate(cluster_dict[cluster_id]):
        try:
            if val != product_id:
                distances.append(1 - spatial.distance.cosine(vector_dict[product_id], vector_dict[val])) #11139524
                
        except:
            pass
    
    index, value = max(enumerate(distances), key=operator.itemgetter(1))
    return cluster_dict[cluster_id][index]
```

```python
from collections import OrderedDict
from tqdm import tqdm
```

```python
def f(c_id, c_no1, c_no2, c_no3, c_no4):
    '''
    c_id = CustomerId
    c_no = ClusterNo
    '''
    a = set(ca[ca['CustomerId']==c_id].drop('CustomerId', axis=1).fillna('None').values[0]) # old product
    
    b = set(di[c_no1].values())
    c = set(di[c_no2].values())
    d = set(di[c_no3].values())
    e = set(di[c_no4].values())
    
    p = list(b-a)    
    q = list(c-a)    
    r = list(d-a)    
    s = list(e-a)
    
    L = [c_id]
    
    for i in [p[:4], q[:3], r[:2], s[:1]]:
        for j in i:
            L.append(j)
            
    
            
#     return OrderedDict((L))
    return np.array(L)
```
